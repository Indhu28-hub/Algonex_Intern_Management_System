```python
# first

def frequency(name):
    freq = {}

    for i in name:
        if i in freq:
            freq[i] += 1
        else:
            freq[i] = 1

    return freq

print(frequency("algonex"))
```

```python
#2 



def duplicates(numbers):
    for i in numbers:
        if numbers.count(i)>1:
            numbers.append(i)
    return numbers
        
print(duplicates([101,102,103,101,104,102]))
```

```python
#3
employee=[
    ("john",50000),
    ("David",70000),
    ("smith",40000),
    ("Alice",90000)
]
highest = employee[0]
lowest = employee[0]
total = 0

for name, salary in employee:
    if salary > highest[1]:
        highest = name, salary
    if salary < lowest[1]:
        lowest = name, salary
    total += salary

average = total / len(employee)
print("highest:",highest)
print("lowest:",lowest)
print("Average:", average)
```

```python
#6
transactions = [1000,-200,500,-100,2000,-500]
credit = 0
debit = 0
for i in transactions:
    if i > 0:
        credit += i
    else:
        debit += abs(i)
balance = credit - debit
print("Total Credit:", credit)
print("Total Debit:", debit)
print("Balance:", balance)
```
